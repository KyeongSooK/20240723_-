## 1. 데이터 로드 및 기본 정보 확인

```python
import pandas as pd

# 파일 경로
activity_sample_path = './activity_sample.csv'
encoded_data_path = './encoded_data.csv'

# 데이터 로드
activity_sample_df = pd.read_csv(activity_sample_path)
encoded_data_df = pd.read_csv(encoded_data_path)
```

```python
# activity_sample.csv 기본 정보
print("activity_sample.csv 기본 정보")
print(activity_sample_df.info())
print(activity_sample_df.describe())

# encoded_data.csv 기본 정보
print("\nencoded_data.csv 기본 정보")
print(encoded_data_df.info())
print(encoded_data_df.describe())
```

## 2. 데이터 시각화

#### 2-1 컬럼 갯수 비교

```python
import matplotlib.pyplot as plt

# 컬럼 갯수 비교
columns_counts = {'activity_sample': len(activity_sample_df.columns),
                  'encoded_data': len(encoded_data_df.columns)}

plt.figure(figsize=(10, 5))
plt.bar(columns_counts.keys(), columns_counts.values(), color=['blue', 'green'])
plt.title('Compare number of columns')
plt.xlabel('dataset')
plt.ylabel('column Count')
plt.show()
```

#### 2-2 컬럼별 결측값 비교 // 하이픈 확인

```python
# 결측값 계산
activity_sample_na = activity_sample_df.isna().sum() # 결측치는 없다.
activity_sample_hyphen = (activity_sample_df == '-').sum() # 하이픈 확인
#encoded_data_na = encoded_data_df.isna().sum()

# 결측값 시각화
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
activity_sample_na.plot(kind='bar', color='blue')
plt.title('activity_sample.csv missing values')
plt.xlabel('Column')
plt.ylabel('missing values Count')

plt.subplot(1, 2, 2)
activity_sample_hyphen.plot(kind='bar', color='blue')
plt.title('activity_sample.csv missing values')
plt.xlabel('Column')
plt.ylabel('hyphen (-) Count')


# plt.subplot(1, 2, 2)
# encoded_data_na.plot(kind='bar', color='green')
# plt.title('encoded_data.csv 결측값')
# plt.xlabel('컬럼')
# plt.ylabel('결측값 수')

plt.tight_layout()
plt.show()
```
