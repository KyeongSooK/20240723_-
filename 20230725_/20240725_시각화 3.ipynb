```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
df_activity = pd.read_csv('./activity_sample.csv')
df_encoded = pd.read_csv('./encoded_data.csv')

# 각 데이터프레임의 정보 출력
print("activity_sample.csv 데이터 정보:")
print(df_activity.info())
print("\nencoded_data.csv 데이터 정보:")
print(df_encoded.info())

# 쉼표를 점으로 대체하여 숫자로 변환 (activity_sample.csv)
df_activity = df_activity.replace({',': '.'}, regex=True).apply(pd.to_numeric, errors='coerce')

# 결측값 처리 (activity_sample.csv)
df_activity.replace('-', np.nan, inplace=True)

# 쉼표를 점으로 대체하여 숫자로 변환 (encoded_data.csv)
df_encoded = df_encoded.replace({',': '.'}, regex=True).apply(pd.to_numeric, errors='coerce')

# 결측값 처리 (encoded_data.csv)
df_encoded.replace('-', np.nan, inplace=True)

# 데이터 프레임 요약 통계량
summary_activity = df_activity.describe()
summary_encoded = df_encoded.describe()

# 컬럼 수 비교
columns_activity = df_activity.shape[1]
columns_encoded = df_encoded.shape[1]

print(f"\nactivity_sample.csv 컬럼 수: {columns_activity}")
print(f"encoded_data.csv 컬럼 수: {columns_encoded}")

```

```python

# 데이터 시각화
plt.figure(figsize=(12, 6))

# activity_sample.csv 데이터 분포 시각화
plt.subplot(1, 2, 1)
sns.histplot(df_activity.values.ravel(), bins=50, kde=True)
plt.title('activity_sample.csv 데이터 분포')

# encoded_data.csv 데이터 분포 시각화
plt.subplot(1, 2, 2)
sns.histplot(df_encoded.values.ravel(), bins=50, kde=True)
plt.title('encoded_data.csv 데이터 분포')

plt.tight_layout()
plt.show()

# 컬럼 수 시각화
plt.figure(figsize=(8, 6))
plt.bar(['activity_sample.csv', 'encoded_data.csv'], [columns_activity, columns_encoded], color=['blue', 'green'])
plt.title('컬럼 수 비교')
plt.ylabel('컬럼 수')
plt.show()

# 데이터 분포 비교
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df_activity)
plt.title('activity_sample.csv 데이터 분포')

plt.subplot(1, 2, 2)
sns.boxplot(data=df_encoded)
plt.title('encoded_data.csv 데이터 분포')

plt.tight_layout()
plt.show()

print(f"\nactivity_sample.csv 컬럼 수: {columns_activity}")
print(f"encoded_data.csv 컬럼 수: {columns_encoded}")
```
